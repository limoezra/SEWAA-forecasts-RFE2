```python
import tensorflow as tf, glob

files = glob.glob(r"E:\CGAN\rfe_tfrecords\2021_1.*.tfrecords")
n = sum(1 for _ in tf.data.TFRecordDataset(files, compression_type="GZIP"))
print(f"2021 records: {n} samples")

for raw in tf.data.TFRecordDataset(files, compression_type="GZIP").take(1):
    ex = tf.train.Example()
    ex.ParseFromString(raw.numpy())
    for key, feat in ex.features.feature.items():
        kind = feat.WhichOneof("kind")
        vals = getattr(feat, kind).value
        print(f"  {key:25s} {kind:12s} len={len(vals)}")
```
